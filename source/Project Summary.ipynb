<h1 id="i" style="font-weight: bold">I. Key Findings</h1>

<h2 id="i_1" style="font-weight: bold">1. Most Important Insights</h2>

* **The "Immunity" of Calorie Monitoring:** The analysis of Question 1 revealed a massive behavioral impact. Among high-risk individuals (those eating high-calorie food or drinking alcohol), those who monitored calories (SCC) had a severe obesity rate of only 1.6%, compared to 34.1% for those who did not. Fisher’s Exact Test confirmed this is statistically significant (p<0.001), suggesting that self-monitoring effectively mitigates the risks of a poor diet,.

* **The "Ordinality" of Obesity:** The analysis in Question 2 proved that treating obesity as a ranked severity problem (Ordinal Classification) is safer for medical risk assessment than standard classification. While standard XGBoost achieved higher raw accuracy (80.1%), the Ordinal Classifier achieved a low Mean Absolute Error (MAE≈0.35), meaning it rarely confuses severe obesity with normal weight, ensuring "structural consistency" in predictions.

* **Personalized vs. General Drivers:** While the general population analysis (Question 6) suggested "Water Intake" is a key differentiator, the patient-specific SHAP analysis (Question 2) revealed that for severe cases (Obesity Type III), Physical Activity (FAF) is the dominant "lever" for improvement. For specific high-risk patients, increasing activity reduced the risk score by 0.40, whereas increasing water had a negligible effect (0.004). This proves that general advice does not always apply to individual high-risk cases.

* **Gender-Specific Obesity Drivers:** Question 3 statistically confirmed that female obesity is driven by "Caloric Imbalance" (low activity), while male obesity is driven by "Environmental Sedentarism" (high automobile use).

* **The "Tech-User" Paradox:** Question 5 debunked the "Digital Sedentary" myth for this dataset. Technology Use (TUE) showed a negative correlation with obesity (r=−0.30), likely acting as a proxy for socioeconomic status. The highest risk group was actually the "Non-Digital Sedentary" group (No Tech + No Exercise).

<h2 id="i_2" style="font-weight: bold">2. Most Interesting or Surprising Discovery</h2>

The "Vegetable Anomaly" (Reverse Causality): The most surprising discovery remains that the Obesity_Type_III group reported the highest vegetable consumption (FCVC≈3.0). This strongly suggests reverse causality—severe obesity drives patients to diet (eat veggies), rather than veggies causing obesity. This highlights the danger of taking self-reported health data at face value without context.

<h1 id="ii" style="font-weight: bold">II. Limitations</h1>

<h2 id="i_1" style="font-weight: bold">1. Dataset Limitations</h2>

* **Extreme Class Imbalance:** The sample size for people who monitor calories was extremely small (n=63 vs 1796 non-monitors). While Fisher's Exact Test handled this mathematically, the scarcity of this behavior limits the model's ability to generalize this finding to the broader population.

* **Self-Reporting Bias:** The "Vegetable Anomaly" confirms significant social desirability bias or "dieting bias" in the survey responses, making nutritional features (FCVC) unreliable predictors of cause.

* **Feature Imbalance:** Critical features like Smoking (SMOKE) and Calorie Monitoring (SCC) were "dead features" with >95% answering "No," limiting the model's ability to learn from these behaviors.

* **Missing Socioeconomic Data:** The analysis of Question 5 implies that "Low Tech Use" correlates with obesity due to socioeconomic factors, but the dataset lacks Income/Education columns to verify this.

<h2 id="ii_2" style="font-weight: bold">2. Analysis Limitations</h2>

* **Cross-Sectional Nature:** The dataset is a snapshot. We cannot determine if the calorie monitors have always been thin or the Vegetable Anamoly that the Obesity_Type_III group having the highest vegetable consumption, or if they successfully lost weight using the method, or we cannot determine if the "Meal Skippers" (Cluster 2 from Question 4) will become obese, or if they are currently losing weight. We see the correlation, but proving the direction of causality is impossible without time-series data.

* **Accuracy vs. Consistency Trade-off:** There was a trade-off between predictive power and medical safety. The Standard XGBoost model was better at distinguishing "Overweight Level I" (Accuracy), but the Ordinal Model was better at preventing dangerous errors (Consistency). The final model had to compromise by using Standard XGBoost for accuracy while using Ordinal logic for the Risk Score calculation.


<h2 id="ii_3" style="font-weight: bold">3. Scope Limitations</h2>

The analysis identified "Meal Skipping" (Cluster 2) and "Sedentary Tech Use" (Cluster 1), but could not assess the mental health state (stress, depression) that often drives these behaviors.

Instead of Frank & Hall decomposition (used in Q2), implement Earth Mover's Distance (EMD) loss in a Neural Network. This penalizes "severe" errors (predicting Type III as Normal) much more heavily than minor errors, potentially solving the accuracy/safety trade-off

<h1 id="iv" style="font-weight: bold">IV. Individual Reflections</h1>

<h2 id="iv_1" style="font-weight: bold">1. Challenges & Difficulties Encountered</h2>

<h3 id="iv_1_a" style="font-weight: bold">a. Lê Võ Nhật Minh</h3>

*   Critically, the target variable (`NObeyesdad`) consists of 7 severity classes with an inherent ordinal structure. Applying a standard multi-class classifier ignores this order, leading to less reliable risk predictions, particularly near classification boundaries. The core conceptual challenge of ordinality was addressed by implementing an "Ordinal Classification approach". This model uses decomposition into binary classifiers and applies "Isotonic Regression" to enforce monotonicity constraints on the predicted probabilities, ensuring that the predicted risk aligns with the severity order.

*   Recognizing significant data quality concerns such as the potential unreliability of subjective features (like `FCVC`—consumption of vegetables) and evidence of self-reported data bias (the "Vegetable Anomaly," where the most obese group reported the highest vegetable consumption). Also, this can be reported by people in diet.

*   The most challenging aspect was achieving a model that balances high predictive accuracy with structural consistency, specifically related to the ordinal nature of the obesity levels. While the standard XGBoost model achieved a higher overall accuracy (85.1%), it lacked the structural guarantee of monotonicity. The Ordinal Classifier, although slightly less accurate overall (78.0%), provided a more reliable measure of risk, evidenced by a lower Mean Absolute Error (MAE=0.354). The challenge lay in the decision to choose the standard XGBoost as the final model because it performed best in predicting the crucial target group ("Overweight Level I").

<h3 id="iv_1_a" style="font-weight: bold">b. Nguyễn Hữu Nghĩa</h3>

- The "Healthy Obese" Paradox: A major analytical hurdle arose when the data indicated that the `Obesity_Type_III` group had the highest Vegetable Consumption score (`FCVC` $\approx 3.0$). This contradicted our initial hypothesis that poor diet correlates linearly with weight. I had to look past the numbers to interpret this as a potential "Reverse Causality" or "Self-Reporting Bias," where patients with severe obesity are likely already on strict diets.

- Dimensionality Reduction Ambiguity: In the Clustering phase (Question 4), visualizing the 4 clusters via PCA presented a difficulty as Cluster 0 and Cluster 2 showed significant spatial overlap. I initially worried the model failed to separate them. I had to pivot from 2D visual reliance to a Feature Heatmap analysis, which successfully revealed that while they looked similar in space, they were behaviorally distinct (one defined by high vegetable intake, the other by meal skipping).

- Imbalanced Categorical Features: During the Categorical Analysis, I found that features I expected to be significant, like Smoking (`SMOKE`) and Calorie Monitoring (`SCC`), had extreme imbalance (>95% "No"). This lack of variance made them statistically "dead features," forcing me to restructure the analysis to focus on more impactful drivers like Transportation (`MTRANS`) and Snacking (`CAEC`).

Learned to use SHAP values and counterfactual simulation to provide model explainability and actionable, personalized prescriptive advice based on modifiable lifestyle changes.

*   I have learned the critical importance of evaluating data features not just by their technical type but by their meaning (e.g., converting frequency-based numerical features to ordinal categories).

*   The project emphasized that structural consistency and reliability are often more crucial than marginal increases in raw accuracy, especially in high-stakes applications like health risk assessment. While simple models may score well, addressing domain-specific structures (like ordinality) is essential for developing models that provide trustworthy risk prediction and decision support. This reinforced the understanding that successful data science requires integrating advanced modeling techniques with methods for explainability and actionability (e.g., SHAP and counterfactuals) to bridge the gap between complex results and practical, personalized recommendations.

- Data Contextualization over Raw Numbers: I learned to question counter-intuitive results. Instead of discarding the finding that "Obese individuals eat the most vegetables," I learned to interpret data through a behavioral lens-understanding that data captures a snapshot of current efforts (dieting), not necessarily the historical habits that caused the obesity.

- Profiling over Classification: By applying Unsupervised Learning (K-Means), I moved beyond simply classifying "How fat is a patient?" to understanding "Why are they fat?". I learned to identify distinct archetypes, discovering that Cluster 1 (Active Tech-Users) needs a completely different medical intervention than Cluster 0 (Sedentary Veggie-Lovers), despite both having weight issues.

While the charts hinted that men and women gain weight differently, applying the Chi-Square Test ($P \approx 0$) allowed me to scientifically prove that Male obesity is driven by Environmental factors (Automobiles), whereas Female obesity is significantly linked to Caloric Imbalance (Sedentary behavior).
